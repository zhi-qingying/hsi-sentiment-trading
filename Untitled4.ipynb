```python
# HSI Trading Strategy with Moving Average Smoothing
# Step 1: Setup and Data Loading

!pip install fpdf

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("Libraries imported successfully")

# Upload and load data
uploaded = files.upload()

# Read the Excel file
hsi_data = pd.read_excel('HSI.xlsx')
hsi_data['Date'] = pd.to_datetime(hsi_data['Date'])
hsi_data = hsi_data.sort_values('Date').set_index('Date')

print(f"Data loaded: {len(hsi_data)} rows")
print(f"Date range: {hsi_data.index.min()} to {hsi_data.index.max()}")

# Check data quality
print(hsi_data[['Open', 'High', 'Low', 'Close', 'Up votes', 'Down votes']].head())
print(f"Missing values in votes: {hsi_data[['Up votes', 'Down votes']].isnull().sum().sum()}")
```

```python
# Step 2: Data Cleaning and Preparation

# Fill missing vote data with neutral (0.5)
hsi_data[['Up votes', 'Down votes']] = hsi_data[['Up votes', 'Down votes']].fillna(0.5)

# Calculate basic sentiment
hsi_data['sentiment_raw'] = hsi_data['Up votes'] - hsi_data['Down votes']

print("Missing data filled with neutral values (0.5)")
print(f"Sentiment range: {hsi_data['sentiment_raw'].min():.3f} to {hsi_data['sentiment_raw'].max():.3f}")

# Show sentiment statistics
print("\nSentiment Statistics:")
print(f"Mean sentiment: {hsi_data['sentiment_raw'].mean():.3f}")
print(f"Positive sentiment days: {len(hsi_data[hsi_data['sentiment_raw'] > 0])}")
print(f"Negative sentiment days: {len(hsi_data[hsi_data['sentiment_raw'] < 0])}")
```

```python
# Step 3: Moving Average Calculation

# Calculate different moving averages
hsi_data['sentiment_ma3'] = hsi_data['sentiment_raw'].rolling(window=3).mean()
hsi_data['sentiment_ma5'] = hsi_data['sentiment_raw'].rolling(window=5).mean()
hsi_data['sentiment_ma10'] = hsi_data['sentiment_raw'].rolling(window=10).mean()

print(" Moving averages calculated:")
print("   - 3-day MA (short-term)")
print("   - 5-day MA (medium-term) - main signal")
print("   - 10-day MA (long-term)")

# Show how moving averages smooth the data
sample_data = hsi_data[['sentiment_raw', 'sentiment_ma3', 'sentiment_ma5', 'sentiment_ma10']].head(10)
print("\nSample of raw vs smoothed sentiment:")
print(sample_data)
```

```python
# Step 4: Visualization

# Create figure with subplots
plt.figure(figsize=(15, 12))

# Plot 1: Price movement
plt.subplot(3, 1, 1)
plt.plot(hsi_data.index, hsi_data['Close'], color='blue', linewidth=1.5)
plt.title('HSI Close Price Over Time')
plt.ylabel('Price')
plt.grid(True, alpha=0.3)

# Plot 2: Raw vs smoothed sentiment
plt.subplot(3, 1, 2)
plt.plot(hsi_data.index, hsi_data['sentiment_raw'], color='gray', alpha=0.6, label='Raw Sentiment')
plt.plot(hsi_data.index, hsi_data['sentiment_ma5'], color='red', linewidth=2, label='5-Day Moving Average')
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
plt.title('Sentiment: Raw vs Smoothed (5-Day MA)')
plt.ylabel('Sentiment Score')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 3: All moving averages
plt.subplot(3, 1, 3)
plt.plot(hsi_data.index, hsi_data['sentiment_ma3'], color='orange', linewidth=1, label='3-Day MA')
plt.plot(hsi_data.index, hsi_data['sentiment_ma5'], color='red', linewidth=2, label='5-Day MA')
plt.plot(hsi_data.index, hsi_data['sentiment_ma10'], color='purple', linewidth=1, label='10-Day MA')
plt.axhline(y=0.1, color='green', linestyle=':', alpha=0.7, label='Buy Threshold (0.1)')
plt.axhline(y=-0.1, color='red', linestyle=':', alpha=0.7, label='Sell Threshold (-0.1)')
plt.title('Moving Average Comparison')
plt.ylabel('Sentiment Score')
plt.xlabel('Date')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(" Charts created successfully")
```

```python
# Step 5: Trading Strategy Implementation

def generate_trading_signals(data):
    """
    Generate buy/sell signals based on smoothed sentiment
    Buy when 5-day MA > 0.1, Sell when < -0.1
    """
    data = data.copy()
    data['signal'] = 0  # 0 = no position, 1 = buy, -1 = sell

    # Buy signal: smoothed sentiment > 0.1
    buy_condition = data['sentiment_ma5'] > 0.1
    data.loc[buy_condition, 'signal'] = 1

    # Sell signal: smoothed sentiment < -0.1
    sell_condition = data['sentiment_ma5'] < -0.1
    data.loc[sell_condition, 'signal'] = -1

    return data

# Apply trading strategy
hsi_data = generate_trading_signals(hsi_data)

# Count signals
buy_days = len(hsi_data[hsi_data['signal'] == 1])
sell_days = len(hsi_data[hsi_data['signal'] == -1])
hold_days = len(hsi_data[hsi_data['signal'] == 0])

print(" Trading Signals Summary:")
print(f"   Buy signals: {buy_days} days ({buy_days/len(hsi_data)*100:.1f}%)")
print(f"   Sell signals: {sell_days} days ({sell_days/len(hsi_data)*100:.1f}%)")
print(f"   Hold days: {hold_days} days ({hold_days/len(hsi_data)*100:.1f}%)")

# Show recent signals
print("\nRecent trading signals:")
recent_signals = hsi_data[['sentiment_ma5', 'signal']].tail(10)
print(recent_signals)
```

```python
# Step 6: Backtesting

# Calculate daily returns
hsi_data['market_return'] = hsi_data['Close'].pct_change()
hsi_data['strategy_return'] = hsi_data['signal'].shift(1) * hsi_data['market_return']

# Remove first row with NaN
backtest_data = hsi_data.dropna(subset=['strategy_return'])

print(f"Backtest period: {len(backtest_data)} trading days")

# Calculate cumulative returns
backtest_data['cumulative_market'] = (1 + backtest_data['market_return']).cumprod()
backtest_data['cumulative_strategy'] = (1 + backtest_data['strategy_return']).cumprod()

print(" Returns calculated successfully")

# Show sample returns
print("\nSample returns data:")
sample_returns = backtest_data[['market_return', 'strategy_return', 'cumulative_market', 'cumulative_strategy']].head()
print(sample_returns)
```

```python
# Step 7: Performance Analysis

# Total returns
total_market_return = backtest_data['cumulative_market'].iloc[-1] - 1
total_strategy_return = backtest_data['cumulative_strategy'].iloc[-1] - 1

# Annual returns
days = len(backtest_data)
annual_market_return = (1 + total_market_return) ** (252/days) - 1
annual_strategy_return = (1 + total_strategy_return) ** (252/days) - 1

# Sharpe ratio
sharpe_market = backtest_data['market_return'].mean() / backtest_data['market_return'].std() * np.sqrt(252)
sharpe_strategy = backtest_data['strategy_return'].mean() / backtest_data['strategy_return'].std() * np.sqrt(252)

# Maximum drawdown
def max_drawdown(returns):
    cumulative = (1 + returns).cumprod()
    peak = cumulative.expanding().max()
    drawdown = (cumulative - peak) / peak
    return drawdown.min()

max_dd_market = max_drawdown(backtest_data['market_return'])
max_dd_strategy = max_drawdown(backtest_data['strategy_return'])

# Win rate
winning_trades = len(backtest_data[backtest_data['strategy_return'] > 0])
total_trades = len(backtest_data[backtest_data['signal'].shift(1) != 0])
win_rate = winning_trades / total_trades * 100 if total_trades > 0 else 0

print(" PERFORMANCE RESULTS:")
print("=" * 50)
print(f"{'Metric':<20} {'Buy & Hold':<12} {'Sentiment Strategy':<15}")
print("=" * 50)
print(f"{'Total Return':<20} {total_market_return:>11.2%} {total_strategy_return:>14.2%}")
print(f"{'Annual Return':<20} {annual_market_return:>11.2%} {annual_strategy_return:>14.2%}")
print(f"{'Sharpe Ratio':<20} {sharpe_market:>11.2f} {sharpe_strategy:>14.2f}")
print(f"{'Max Drawdown':<20} {max_dd_market:>11.2%} {max_dd_strategy:>14.2%}")
print(f"{'Win Rate':<20} {'N/A':>11} {win_rate:>14.1f}%")
```

```python
# Step 8: Results Visualization

# Create performance charts
plt.figure(figsize=(15, 10))

# Chart 1: Cumulative returns
plt.subplot(2, 1, 1)
plt.plot(backtest_data.index, backtest_data['cumulative_market'],
         label='Buy & Hold', linewidth=2, color='blue')
plt.plot(backtest_data.index, backtest_data['cumulative_strategy'],
         label='Sentiment Strategy (5-day MA)', linewidth=2, color='red')
plt.title('Performance: Sentiment Strategy vs Buy & Hold')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True, alpha=0.3)

# Chart 2: Trading signals on price
plt.subplot(2, 1, 2)
plt.plot(hsi_data.index, hsi_data['Close'], color='black', linewidth=1, label='HSI Price')

# Mark buy signals
buy_signals = hsi_data[hsi_data['signal'] == 1]
plt.scatter(buy_signals.index, buy_signals['Close'],
           color='green', marker='^', s=40, label='Buy', alpha=0.7)

# Mark sell signals
sell_signals = hsi_data[hsi_data['signal'] == -1]
plt.scatter(sell_signals.index, sell_signals['Close'],
           color='red', marker='v', s=40, label='Sell', alpha=0.7)

plt.title('Trading Signals on Price Chart')
plt.ylabel('Price')
plt.xlabel('Date')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(" Results visualization completed")
```

```python
# Step 9: Export Results

# Create summary table
results_summary = pd.DataFrame({
    'Metric': ['Total Return', 'Annual Return', 'Sharpe Ratio', 'Max Drawdown', 'Win Rate'],
    'Buy_Hold': [
        f"{total_market_return:.2%}",
        f"{annual_market_return:.2%}",
        f"{sharpe_market:.2f}",
        f"{max_dd_market:.2%}",
        'N/A'
    ],
    'Sentiment_Strategy': [
        f"{total_strategy_return:.2%}",
        f"{annual_strategy_return:.2%}",
        f"{sharpe_strategy:.2f}",
        f"{max_dd_strategy:.2%}",
        f"{win_rate:.1f}%"
    ]
})

print("Final Results Summary:")
print(results_summary)

# Save to files
results_summary.to_csv('hsi_sentiment_strategy_results.csv', index=False)
backtest_data.to_csv('hsi_backtest_detailed.csv')

print("\n Files saved:")
print("   - hsi_sentiment_strategy_results.csv (summary)")
print("   - hsi_backtest_detailed.csv (detailed data)")

print("\n PROJECT COMPLETED SUCCESSFULLY!")
print("The moving average smoothing has been applied to improve trading decisions.")
```
